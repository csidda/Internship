```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import ElementClickInterceptedException,ElementNotInteractableException,NoSuchElementException
```

Q1 Scrape the details of most viewed videos on YouTube from Wikipedia

```python
driver = webdriver.Chrome()
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
driver.maximize_window()
time.sleep(3)
names = []
artist = []
views = []
date = []
name_tag = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
artist_tag = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
view_tag = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
date_tag = driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')

for i in name_tag:
    names.append(i.text)
    
for i in artist_tag:
    artist.append(i.text)
    
for i in view_tag:
    views.append(i.text)
    
for i in date_tag:
    date.append(i.text)
    
driver.close()   

df = pd.DataFrame({"Name":names,"Artist":artist,"Upload date":date,"Viwes":views})
df
```

Q2 Scrape the details team India‚Äôs international fixtures from bcci.tv

```python
driver = webdriver.Chrome()
url = "https://www.bcci.tv/"
driver.get(url)
driver.maximize_window()
fixtures = driver.find_element(By.XPATH,'//div[@id="pills-international"]/ul/div[1]/a[2]')
fixtures.click()
time.sleep(5)
    
series = []
place = []
date = []
time = []

series_tag = driver.find_elements(By.XPATH,'//div[@class="match-card-top"]/div[2]/h5')
place_tag = driver.find_elements(By.XPATH,'//div[@class="match-card-top"]/div[2]/div[3]')
date_tag = driver.find_elements(By.XPATH,'//div[@class="match-card-top"]/div[2]/div[1]/div[1]')
time_tag = driver.find_elements(By.XPATH,'//div[@class="match-card-top"]/div[2]/div[1]/div[2]')

for i in series_tag:
    series.append(i.text)
    
for i in place_tag:
    place.append(i.text)

for i in date_tag:
    date.append(i.text)

for i in time_tag:
    time.append(i.text)
driver.close()

df = pd.DataFrame({"Series":series,"Place":place,"Date":date,"Time":time})
df
```

Q3 Scrape the details of State-wise GDP of India from statisticstime.com. 

```python
driver = webdriver.Chrome()
url = "http://statisticstimes.com/"
driver.get(url)
driver.maximize_window()
time.sleep(5)
economy_tag = driver.find_element(By.XPATH,'//div[@class="navbar"]/div[2]/button/i')
economy_tag.click()
time.sleep(5)
indian_tag = economy_tag.find_element(By.XPATH,'//div[@class="navbar"]/div[2]/div/a[3]')
indian_tag.click()
time.sleep(5)
state_wise_gdp = driver.find_element(By.XPATH,'//a[@href="india/indian-states-gdp.php"]')
state_wise_gdp.click()
time.sleep(20)
rank = []
state = []
gsdp = []
gsdp1 = []
share = []
gdp = []

rank_tag = driver.find_elements(By.XPATH,'//div[@id="table_id_wrapper"]/table/tbody/tr/td[1]')
state_tag = driver.find_elements(By.XPATH,'//div[@id="table_id_wrapper"]/table/tbody/tr/td[2]')
gsdp_tag = driver.find_elements(By.XPATH,'//div[@id="table_id_wrapper"]/table/tbody/tr/td[5]')
gsdp1_tag = driver.find_elements(By.XPATH,'//div[@id="table_id_wrapper"]/table/tbody/tr/td[10]')
share_tag = driver.find_elements(By.XPATH,'//div[@id="table_id_wrapper"]/table/tbody/tr/td[6]')
gdp_tag = driver.find_elements(By.XPATH,'//div[@id="table_id_wrapper"]/table/tbody/tr/td[7]')
for i in rank_tag:
    rank.append(i.text)

for i in state_tag:
    state.append(i.text) 
    
for i in gsdp_tag:
    gsdp.append(i.text)
    
for i in gsdp1_tag:
    gsdp1.append(i.text)

for i in share_tag:
    share.append(i.text)
    
for i in gdp_tag:
    gdp.append(i.text)
    
driver.close()
df = pd.DataFrame({"Rank":rank,"State":state,"GSDP (21-22)-at current prices":gsdp,"GSDP (21-22) at 2011-12 price":gsdp1,"Share (21-22)":share,"GDP ($billion)":gdp})
df
```

Q4 Scrape the details of trending repositories on Github.com. 

```python
driver = webdriver.Chrome()
url = "https://github.com/"
driver.get(url)
driver.maximize_window()
time.sleep(5)

search_options = driver.find_element(By.XPATH,'//nav[@class="HeaderMenu-nav"]/ul/li[4]/button')
search_options.click()
time.sleep(3)
trading_option = search_options.find_element(By.XPATH,'//a[@href="https://github.com/trending"]')
trading_option.click()
time.sleep(5)

repository_title = []
repository_description = []
language_used = []
count = []

title_tag = driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
description_tag = driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
language_tag = driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/span[1]')
count_tag = driver.find_elements(By.XPATH,'//div[@class="f6 color-fg-muted mt-2"]/a[2]')

for i in title_tag:
    repository_title.append(i.text)
    
for i in description_tag:
    repository_description.append(i.text)
    
for i in language_tag:
    language_used.append(i.text)
    
for i in count_tag:
    count.append(i.text)
 
driver.close()

df = pd.DataFrame({"Repository title":repository_title,"Repository description":repository_description,"Contributors count":count,"Language used":language_used})
df
```

Q5 Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/ 

```python
driver = webdriver.Chrome()
driver.get("https:/www.billboard.com/")
driver.maximize_window()
time.sleep(2)

try:
    chart_option = driver.find_element(By.XPATH,'//div[@class="lrv-u-flex lrv-u-flex-basis-30p lrv-u-justify-content-center lrv-u-align-items-center"]/div/nav/ul/li[1]/a')
    chart_option.click()
except ElementNotInteractableException:
    chart_option = driver.find_element(By.XPATH,'//div[@class="lrv-u-flex lrv-u-flex-basis-30p lrv-u-justify-content-center lrv-u-align-items-center"]/div/nav/ul/li[1]/a')
    driver.get(chart_option.get_attribute('href'))
    time.sleep(10)
    
hot_100 = driver.find_element(By.XPATH,'//div[@class="main-sub-menu-container-esp js-hide-when-sticky lrv-u-flex lrv-u-background-color-white u-height-45"]/div/nav/ul/li[1]')
hot_100.click()
time.sleep(5)

song_name = []
artist_name = []
last_week_rank = []
peak_rank = []
week_board = []

song_tag = driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[1]/h3')
artist_tag = driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[1]/span')
last_week_rank_tag = driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[4]/span')
peak_rank_tag = driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[5]/span')
week_board_tag = driver.find_elements(By.XPATH,'//li[@class="lrv-u-width-100p"]/ul/li[6]/span')

for i in song_tag:
    song_name.append(i.text)
    
for i in artist_tag:
    artist_name.append(i.text)

for i in last_week_rank_tag:
    last_week_rank.append(i.text)
    
for i in peak_rank_tag:
    peak_rank.append(i.text)

for i in week_board_tag:
    week_board.append(i.text)

driver.close()

df = pd.DataFrame({"Song Name":song_name,"Artist Name":artist_name,"Last Week Rank":last_week_rank,"Peak Rank":peak_rank,"Weeks on board":week_board})
df
```

Q8 Details of Datasets from UCI machine learning repositories. 

```python
driver = webdriver.Chrome()
url = "https://archive.ics.uci.edu"
driver.get(url)
driver.maximize_window()
time.sleep(5)

search = driver.find_element(By.XPATH,'//input[@class="input join-item input-bordered input-sm w-full"]')
search.send_keys("all dataset")
search_button = driver.find_element(By.XPATH,'//button[@class="btn btn-primary join-item btn-sm"]')
search_button.click()
time.sleep(5)

try:
    expand = driver.find_element(By.XPATH,'//label[@class="btn btn-primary btn-sm flex gap-2 rounded-full"]')
    expand.click()
    time.sleep(5)
except NoSuchElementException:
    expand = driver.find_element(By.XPATH,'//label[@class="btn btn-primary btn-sm flex gap-2 rounded-full"]/span[2]/span[1]')
    expand.click()
    time.sleep(3)

dataset_name = []
data_type = []
task = []
number_instance = []
number_attribute = []
attribute_type = []
year = []

datasetname_tag = driver.find_elements(By.XPATH,'//h2[@class="truncate text-primary"]/a')
datatype_tag = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
task_tag = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
instance_tag = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
attribute_tag = driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
attributetype_tag = driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[2]')
year_tag = driver.find_elements(By.XPATH,'//div[@class="grid grid-cols-8 overflow-x-auto"]/table/tbody/tr/td[3]')

for i in datasetname_tag:
    dataset_name.append(i.text)
for i in datatype_tag:
    data_type.append(i.text)
for i in task_tag:
    task.append(i.text)
for i in instance_tag:
    number_instance.append(i.text)
for i in attribute_tag:
    number_attribute.append(i.text)
for i in attributetype_tag:
    attribute_type.append(i.text)
for i in year_tag:
    year.append(i.text)

driver.close()
df = pd.DataFrame({"Dataset Name":dataset_name,"Data Type":data_type,"Task":task,"Attribute Type":attribute_type,"Number of Instance":number_instance,"Number of Attribute":number_attribute,"Year":year})
df
```
