```python
import pandas as pd
import selenium
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.common.exceptions import ElementClickInterceptedException,NoSuchElementException
```

Q2       

```python
brand_name = []
product_name = []
prices = []
product_url = []
start = 0
end = 3
for page in range(start,end):
    brand_tag = driver.find_elements(By.XPATH,'//h2[@class="a-size-mini s-line-clamp-1"]')
    product_tag = driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
    price_tag = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-no-hover s-underline-text s-underline-link-text s-link-style a-text-normal"]/span[1]')
    url = driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    
    for i in brand_tag:
        brand_name.append(i.text)
    for i in product_tag:
        product_name.append(i.text)
    for i in price_tag:
        price = i.text
        prices.append(price.replace("₹",""))
   
    
    for i in url:
        product_url.append(i.get_attribute('href'))
    next_button = driver.find_element(By.XPATH,'//*[text()="Next"]')
    next_button.click()
    time.sleep(5) 

return_exchange = []
for i  in product_url:
    driver.get(i)
    time.sleep(5)
    try:
        exchange = driver.find_element(By.XPATH,'//span[@class="a-declarative"]/div[2]/span')
        return_exchange.append(exchange.text)
    except NoSuchElementException:
        return_exchange.append('-')
driver.close()
        
df = pd.DataFrame({"Brand Name":brand_name,"Name of the product":product_name,"Price":prices,"Return/Exchange":return_exchange,"Product URL":product_url})
df
```

```python
#to save dataframe in csv
df2 = df.to_csv("product_details.csv",index=False)
```

```python
driver = webdriver.Chrome()
driver.get("https://www.forbes.com")
driver.maximize_window()
time.sleep(5)

path = driver.find_element(By.XPATH,'//div[@class="dN9h-rXs"]/div')
path.click()
time.sleep(5)

path_select = driver.find_element(By.XPATH,'//li[@class="QpfB0oZe ocPoioVz primary"][1]/div/span')
path_select.click()
time.sleep(2)

path_selector = driver.find_element(By.XPATH,'//div[@class="A6l1kjob G-fdXOrK"]/div[3]/ul/li[1]/a')
path_selector.click()
time.sleep(5)

rank = []
name = []
networth = []
industry = []

page_start = 1
page_end = 15
for page in range(page_start,page_end):
    rank_tag = driver.find_elements(By.XPATH,'//div[@class="Table_rank__X4MKf"]/div[1]')
    for i in rank_tag:
        ranks = i.text
        rank.append(ranks.replace('.',''))


    name_tag = driver.find_elements(By.XPATH,'//div[@class="Table_personName__Bus2E"]')
    for i in name_tag:
        name.append(i.text)    


    networth_tag = driver.find_elements(By.XPATH,'//div[@class="Table_finalWorth__UZA6k"]/span')
    for i in networth_tag:
        networths = i.text
        networth.append(networths.replace("$",""))


    industry_tag= driver.find_elements(By.XPATH,'//div[@role="cell"][4]')
    for i in industry_tag:
        industry.append(i.text)
        
    try:
        next_button = driver.find_element(By.XPATH,'//button[@class="t7aCs7DQ _0BYTGi5u TNAO5JyH bMlalnxk _5nHoWK7c"][2]')
        next_button.click()
    except ElementClickInterceptedException as e:
        next_button = driver.find_element(By.XPATH,'//button[@class="t7aCs7DQ _0BYTGi5u TNAO5JyH bMlalnxk _5nHoWK7c"][1]')
        next_button.click()
        
driver.close()    
df = pd.DataFrame({"RANK":rank,"NAME":name,"NET WORTH":networth,"INDUSTRY":industry})
df
```

```python
driver = webdriver.Chrome()
driver.get("https://www.youtube.com/watch?v=cE-Ej1ycXtk")
driver.maximize_window()


comments = []
upvote = []
time = []
comment_tag = driver.find_elements(By.XPATH,'//yt-attributed-string[@id="content-text"]')
vote_tag = driver.find_elements(By.XPATH,'//span[@id="vote-count-middle"]')
time_tag = driver.find_elements(By.XPATH,'//span[@class="style-scope ytd-comment-view-model"]/a')
for i in comment_tag:
    comments.append(i.text)
    
for i in vote_tag:
    vote = i.text
    upvote.append(vote)
    
for i in time_tag:
    time.append(i.text)
driver.close()
    
df = pd.DataFrame({"Comments":comments,"UpVote":upvote,"Time":time})
df
```

```python
driver = webdriver.Chrome()
driver.get("https://www.hostelworld.com/")
driver.maximize_window()

location = driver.find_element(By.XPATH,'/html/body/div[3]/div/div[3]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div/div/div[2]/label/input')
location.send_keys("London")
auto_popup = driver.find_elements(By.XPATH,'/html/body/div[3]/div/div[3]/main/header/div/div[2]/div[1]/div[1]/div/div[1]/div[2]/div/ul/li[2]/button')
auto_popup.click()
search = driver.find_element(By.XPATH,'//button[@class="btn-content large-button"]')
search.click()

hostel_name = []
distance_from_city = []
rating = []
review = []


name_tag = driver.find_elements(By.XPATH,'//div[@class="property-name"]/span')
distance_tag = driver.find_elements(By.XPATH,'//div[@class="property-distance"]/span[2]')
rating_tag = driver.find_elements(By.XPATH,'//div[@class="rating medium theme-null"]/span[1]')
review_tag = driver.find_elements(By.XPATH,'//div[@class="rating medium theme-null"]/span[3]')

for i in name_tag:
    hostel_name.append(i.text)
    
for i in distance_tag:
    distance = i.text
    distance_from_city.append(distance.replace("- ",""))
    
for i in rating_tag:
    rating.append(i.text)

for i in review_tag:
    reviews = i.text.replace("(","")
    review.append(reviews.replace(")",""))
    

driver.close()    
df = pd.DataFrame({"Hostel Name":hostel_name,"Distance":distance,"Rating":rating})
df
```
