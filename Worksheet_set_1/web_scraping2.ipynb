1. In this question you have to scrape data using the filters available on the webpage You have to use the location and
salary filter. 

```python
driver = webdriver.Chrome()
driver.get("https://www.naukri.com/")
driver.maximize_window()
time.sleep(5)
job_titles = []
job_locations = []
company_names =[]
experience_required = []

designation = driver.find_element(By.XPATH,'//input[@placeholder="Enter skills / designations / companies"]')
designation.send_keys("Data scientist")
search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()

time.sleep(5)
location = driver.find_element(By.XPATH,'//label[@class="styles_chkLbl__n2x09" and @for="chk-Delhi / NCR-cityTypeGid-"]')
location.click()
time.sleep(5)

salary = driver.find_element(By.XPATH,'//label[@class="styles_chkLbl__n2x09" and @for="chk-3-6 Lakhs-ctcFilter-"]')
salary.click()
time.sleep(5)


title_tag = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[1]/a')
for i in title_tag[:10]:
    job_titles.append(i.text)
    

location_tag = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[3]/div/span[3]')
for i in location_tag[:10]:
    job_locations.append(i.text) 
    

company_tag = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[2]/span/a[1]')
for i in company_tag[:10]:
    company_names.append(i.text)
    

experience_tag = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div[3]/div/span[1]')
for i in experience_tag[:10]:
    experience_required.append(i.text)
    
driver.close()
    
df = pd.DataFrame({"Job Title":job_titles,"Job Location":job_locations,"Company Name":company_names,"Experience Required":experience_required})
df
```

2. Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the
job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
driver = webdriver.Chrome()
driver.get("Https://www.shine.com/")
driver.maximize_window()
driver.implicitly_wait(1000)
popup = driver.find_element(By.XPATH,'//div[@class="modal_modalWrapper__DZCQo "]/button')
popup.click()

job_titles = []
locations = []
company_name = []
experience_required = []


skills = driver.find_element(By.XPATH,'//input[@placeholder="Job title, skills"]')
skills.send_keys("Data Analyst")
submit = driver.find_element(By.XPATH,'//span[@class="btn"]/i')
submit.click()

job_title = driver.find_element(By.ID,"id_q")
job_title.send_keys("Data Analyst")
location = driver.find_element(By.ID,'id_loc')
location.send_keys("Bangalore")
search = driver.find_element(By.XPATH,'//button[@type="submit"]')
search.click()

driver.implicitly_wait(1000)
popup1 = driver.find_element(By.XPATH,'//div[@class="modal_modalWrapper__DZCQo "]/button')
popup1.click()


title_tag = driver.find_elements(By.XPATH,'//strong[@class="jobCard_pReplaceH2__xWmHg"]')
for i in title_tag[:10]:
    job_titles.append(i.text)
    

location_tag = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_lists__fdnsc"]/div[1]')
for i in location_tag[:10]:
    locations.append(i.text)
    

name_tag = driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in name_tag[:10]:
    company_name.append(i.text)
    

experience_tag = driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tag[:10]:
    experience_required.append(i.text)
    
driver.close()
    
df = pd.DataFrame({"Job titles":job_titles,"Job location":locations,"Company_name":company_name,"Experience_required":experience_required})
df
```

3. Scrape 100 reviews data from flipkart.com for iphone11 phone.

```python
driver = webdriver.Chrome()
url ="https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART"
driver.get(url)
driver.maximize_window()

ratings = []
review_summary = []
full_review = []
start_page = 1
end_page = 11
for page in range(start_page,end_page):
    rating_tag = driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    
    review_tag = driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
    
    full_review_tag = driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]/div')
    
    for i in rating_tag:
        ratings.append(i.text)
    
    for i in review_tag:
        review_summary.append(i.text)
    
    for i in full_review_tag:
        full_review.append(i.text)
    
    next_button = driver.find_element(By.XPATH,'//*[text()="Next"]')
    next_button.click()
    time.sleep(5)

driver.close()

df = pd.DataFrame({"Ratings":ratings,"Review Summary":review_summary,"Full Review":full_review})
df    
```

4. Scrape data forfirst 100 sneakers you find whenyouvisitflipkart.com and search for “sneakers” inthe search
field.

```python
driver = webdriver.Chrome()
driver.get("https://www.flipkart.com/")
driver.maximize_window()

search_product = driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
search_product.send_keys("Sneakers")
search = driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
search.click()

brands = []
prices = []
product_description = []
page_start = 1
page_end = 4
for page in range(page_start,page_end):
    brand_tag = driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    price_tag = driver.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')
    product_description_tag = driver.find_elements(By.XPATH,'//div[@class="hCKiGj"]/a[1]')
    for i in brand_tag:
        brands.append(i.text)
    for i in price_tag:
        prices.append(i.text)
    for i in product_description_tag:
        product_description.append(i.text)
    next_button = driver.find_element(By.XPATH,'//*[text()="Next"]')
    next_button.click()
    time.sleep(5)
    
Brand = []
for i in brands[:100]:
    Brand.append(i)
Price = []
for i in prices[:100]:
    Price.append(i)
Description = []
for i in product_description[:100]:
    Description.append(i)

driver.close()

df = pd.DataFrame({"Brand":Brand,"Price":Price,"Product Description":Description})
df
```

5. Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU
Type filter to “Intel Core i7”.
Aftersetting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
a.Title
b.Ratings
c.Price


```python
driver = webdriver.Chrome()
driver.get("https://www.amazon.in/")
driver.maximize_window()

search_tag = driver.find_element(By.ID,"twotabsearchtextbox")
search_tag.send_keys("Laptop")
search_button = driver.find_element(By.ID,"nav-search-submit-button")
search_button.click()
time.sleep(5)

cpu_type = driver.find_element(By.XPATH,'//li[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/div')
cpu_type.click()
time.sleep(5)

titles = []
title_tag = driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
for i in title_tag[:10]:
    titles.append(i.text)

ratings = []
rating_tag = driver.find_elements(By.XPATH,'//div[@data-cy="reviews-block"]/div/span[2]/a')
for i in rating_tag[:10]:
    ratings.append(i.text)
    
prices = []
price_tag = driver.find_elements(By.XPATH,'//div[@class="a-section a-spacing-none a-spacing-top-micro puis-price-instructions-style"]/div/div/a/span')
for i in price_tag[:10]:
    prices.append(i.text.replace('₹',""))
    
driver.close()

df = pd.DataFrame({"Title":titles,"Rating":ratings,"Price":prices})
df
```

6. Write a python program to scrape data for Top 1000 Quotes of All Time.

```python
driver = webdriver.Chrome()
driver.get("https://www.azquotes.com/")
driver.maximize_window()

top_quote = driver.find_element(By.XPATH,'//a[@href="/top_quotes.html"]')
top_quote.click()

quote = []
author = []
type_of_quote = []
page_start = 0
page_end = 9
for page in range(page_start,page_end):
    quote_tag = driver.find_elements(By.XPATH,'//a[@class="title"]')
    author_tag = driver.find_elements(By.XPATH,'//div[@class="author"]')
    type_of_quote_tag = driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in quote_tag:
        quote.append(i.text)
    for i in author_tag:
        author.append(i.text)
    for i in type_of_quote_tag:
        type_of_quote.append(i.text)
    next_button = driver.find_element(By.XPATH,"//li[@class='next']")
    next_button.click()
    time.sleep(5)
    
driver.close()

df= pd.DataFrame({"Quote":quote,"Author":author,"Type of Quote":type_of_quote})
df
```

7. Write a python program to display list of respected former Prime Ministers of India (i.e. Name,
Born-Dead, Term of office, Remarks) from https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1

```python
driver = webdriver.Chrome()
driver.get("https://www.jagranjosh.com/general-knowledge/list-ofall-prime-ministers-of-india-1473165149-1")
driver.maximize_window()

Name =[]
name = driver.find_elements(By.XPATH,'//table[@dir="ltr" and @border ="1"]/tbody/tr/td[2]')
for i in name[1:]:
    Name.append(i.text)
    
age_when_assumed_office  = []
age = driver.find_elements(By.XPATH,'//table[@dir="ltr" and @border ="1"]/tbody/tr/td[3]')
for i in age[1:]:
    age_when_assumed_office.append(i.text)
    
start_term_of_office = []
office = driver.find_elements(By.XPATH,'//table[@dir="ltr" and @border ="1"]/tbody/tr/td[4]')
for i in office[1:]:
    start_term_of_office.append(i.text)
    
end_term_of_office1 = []
offices = driver.find_elements(By.XPATH,'//table[@dir="ltr" and @border ="1"]/tbody/tr/td[5]')
for i in offices[1:]:
    end_term_of_office1.append(i.text)
    
durations = []
duration = driver.find_elements(By.XPATH,'//table[@dir="ltr" and @border ="1"]/tbody/tr/td[6]')
for i in duration:
    durations.append(i.text)

driver.close()

df = pd.DataFrame({"Name":Name,"Age when assumed office":age_when_assumed_office,"Start Term Of Office":start_term_of_office,"End Term Of Office":end_term_of_office1,"Duration":durations})
df
```

8. Write a python program to display list of 50 Most expensive cars in the world
(i.e. Car name and Price)

```python
driver = webdriver.Chrome()
driver.get("https://www.motor1.com/")
driver.maximize_window()

search = driver.find_element(By.ID,"search_input")
search.send_keys("50 most expensive car")
search_button = driver.find_element(By.XPATH,'//button[@class="m1-search-panel-button m1-search-form-button-animate icon-search-svg"]')
search_button.click()

time.sleep(5)
car = driver.find_element(By.XPATH,'/html/body/div[9]/div[9]/div/div[1]/div/div/div[1]/div/div[1]/h3/a')
car.click()
time.sleep(3)

car_names = []
car_prices = []
name_tag = driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
price_tag = driver.find_elements(By.XPATH,'//p/strong')
for i in name_tag[:50]:
    car_names.append(i.text)
for i in price_tag[:50]:    
    car_prices.append(i.text.replace("Price: ",""))
car = []
for i in car_prices:
    car.append(i.replace("$",""))

driver.close()

df = pd.DataFrame({"Car Name":car_names,"Car Price":car})
df
```
