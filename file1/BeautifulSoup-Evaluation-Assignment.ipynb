4. Write a python program to scrape first 10 product details which include product name , price , Image URL from
https://www.bewakoof.com/bestseller?sort=popular.

```python
url = "https://shop.bewakoof.com/bestseller?sort=popular"
page = requests.get(url)
soup = BeautifulSoup(page.content)

product_name=[]
price=[]
image=[]
product_names=soup.find_all('div',class_="product-card_product_container__NkgVJ")
for i in product_names[:10]:
    product_name.append(i.text)

prices = soup.find_all('span',class_="sc-a6c4ca6a-0 ccQTLI sc-8020ee44-11 kbHGvj")
for i in prices[:10]:
    price.append(i.text)

images=soup.find_all('figure',class_="sc-8020ee44-0 eMFuRs")
for i in images[:10]:
    i=i.find('img')['src']
    image.append(i)

df = pd.DataFrame({"Product_name":product_name,"Price":price,"Image":image})
df
```

2. Write a python program to scrape details of all the posts from https://www.patreon.com/coreyms .Scrape the
heading, date, content and the likes for the video from the link for the youtube video from the post.

```python
driver = webdriver.Chrome()
driver.get("https://www.patreon.com/coreyms")
driver.maximize_window()
headings = []
heading_tag = driver.find_elements(By.XPATH,'//span[@class="sc-1cvoi1y-0 hxhWXn"]')
for i in heading_tag:
    heading = i.text
    headings.append(heading)

dates = []
date_tag = driver.find_elements(By.XPATH,'//span[@id="track-click"]')
for i in date_tag:
    date = i.text
    dates.append(date)

contents = []
content_tag = driver.find_elements(By.XPATH,'//div[@class="sc-1cq1psq-0 kWkNqn sc-pn81yk-0 dBfLIG"]')
for i in content_tag:
    content = i.text
    contents.append(content)

likes =[]
like_tag = driver.find_elements(By.XPATH,'//div[@class="sc-jrQzAO knZGTZ"]')
for i in like_tag:
    like= i.text
    likes.append(like)

driver.close()

df = pd.DataFrame({"Headings":headings,"Dates":dates,"Contents":contents,"Likes":likes})
df
```

3. Write a python program to scrape house details from mentioned URL. It should include house title, location,
area, EMI and price from https://www.nobroker.in/ .Enter three localities which are Indira Nagar, Jayanagar,
Rajaji Nagar.

```python
driver = webdriver.Chrome()
driver.get("https://www.nobroker.in/")
driver.maximize_window()

localities = driver.find_element(By.XPATH,'//input[@id="listPageSearchLocality"]')

items = ["Indiranagar","Jayanagar","Rajajinagar"]
for i in items:
    localities.send_keys(i)
    driver.implicitly_wait(1000)
    localitie = driver.find_element(By.XPATH,'//div[@id="autocomplete-dropdown-container"]/div/div[1]')
    localitie.click()
   
search = driver.find_element(By.XPATH,'//button[@class="prop-search-button flex items-center justify-center btn btn-primary btn-lg"]')
search.click()

house_title = []
house_tags = driver.find_elements(By.XPATH,'//h2[@class="heading-6 flex items-center font-semi-bold m-0"]/a')
for i in house_tags:
    house = i.text
    house_title.append(house)

location = []
location_tags = driver.find_elements(By.XPATH,'//div[@class="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0.1p po:max-w-95"]')
for i in location_tags:
    locations = i.text
    location.append(locations)

price = []
price_tag = driver.find_elements(By.ID,"minimumRent")
for i in price_tag:
    prices = i.text
    price.append(prices.replace("+",""))
    
driver.close()
df =pd.DataFrame({"House_titles":house_title,"Locations":location,"Prices":price})
df
```
